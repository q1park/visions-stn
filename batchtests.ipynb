## Tensorflow data loading test

```python
def tfloadtest(loader, batchsize = 100, norm = False, onehot = False, shuffle = True):
    loader.tfload(batchsize = batchsize, norm = norm, onehot = onehot, shuffle = shuffle)
```

```python
imgloader = imagefolder(basedir, [120, 120], 3)
imgloader.loadpaths('train')

%timeit -n 100 -r 10 tfloadtest(imgloader)
objgraph.show_growth(limit=5)
```

```python
def tfbatchruntest(tfdict, testprint = False):
    train_dataset = tfdict['train']

    for epoch in range(3):
        for i, itrainbatch in enumerate(train_dataset):
            
            if testprint and epoch == 1 and i < 3:
                print(itrainbatch[0].numpy().shape, itrainbatch[1].numpy().shape)
            pass
```

```python
imgloader = imagefolder(basedir, [120, 120], 3)
imgloader.loadpaths('train')
imgloader.tfload(batchsize = 100, norm = False, onehot = False, shuffle = True)

%timeit -n 3 -r 3 tfbatchruntest(imgloader.tfdict)
objgraph.show_growth(limit=5)
```

```python
tfbatchruntest(imgloader.tfdict, testprint = True)
```

## Numpy data loading test

```python
def nploadtest(loader, norm = False, onehot = False, ncores = 4):
    loader.npload()
```

```python
imgloader = imagefolder(basedir, [120, 120], 3)
imgloader.loadpaths('train')

%timeit -n 3 -r 3 nploadtest(imgloader)
objgraph.show_growth(limit=5)
```

```python
def npbatchruntest(npdict, batchid, testprint = False):
    train_dataset = npdict['train']

    for epoch in range(3):
        for i, ibatch in enumerate(batchid):
            x_trainbatch = train_dataset[0][ibatch]
            y_trainbatch = train_dataset[1][ibatch]
            
            if testprint and epoch == 1 and i < 3:
                print(x_trainbatch.shape, y_trainbatch.shape)
            pass
```

```python
imgloader = imagefolder(basedir, [120, 120], 3)
imgloader.loadpaths('train')
imgloader.npload(norm = False, onehot = True, ncores = 4)
nptestbatchid = imgloader.npbatchid(xtype = 'train', batchsize = 100)

%timeit -n 3 -r 3 npbatchruntest(imgloader.npdict, nptestbatchid)
objgraph.show_growth(limit=5)
```

```python
npbatchruntest(imgloader.npdict, nptestbatchid, testprint = True)
```

## Image quality test

```python
def plot_image(image):
    # Create figure with 3x3 sub-plots.
    fig, ax = plt.subplots(1, 1)
    
    if image.shape[2] == 1:
        image = image[:, :, 0]
    ax.imshow(image.astype(int), cmap = 'gray')
    plt.show()
```

```python
imgloader = imagefolder(basedir, [120, 120], 3)
imgloader.loadpaths('test', classes = ['1'])
imgloader.tfload(batchsize = 1, norm = False, onehot = False, shuffle = False)
imgloader.npload(norm = False, onehot = False, ncores = 4)
```

```python
tfdata = imgloader.tfdict['test']
tfimg = list(enumerate(tfdata))[0][1][0].numpy()[0, :, :, :]
plot_image(tfimg)

npimg = imgloader.npdict['test'][0][0, :, :, :]
plot_image(npimg)
```

```python
imgloader = imagefolder(basedir, [120, 120], 3)
imgloader.loadpaths('test', classes = ['1'])
imgloader.tfload(batchsize = 1, norm = True, onehot = False, shuffle = False)
imgloader.npload(norm = True, onehot = False, ncores = 4)
```

```python
tfdata = imgloader.tfdict['test']
tfimg = list(enumerate(tfdata))[0][1][0].numpy()[0, :, :, :]
plot_image(200.*(tfimg + 1.) )

npimg = imgloader.npdict['test'][0][0, :, :, :]
plot_image(200.*(npimg + 1.) )
```
